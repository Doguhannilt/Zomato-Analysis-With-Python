```python
df = pd.read_excel(r"C:\Users\doguy\Desktop\Zomato\Country-Code.xlsx")
df
```

```python
final_df = pd.merge(data,df,on = 'Country Code' , how = 'left')
```

```python
## To check Data Types

final_df.dtypes
```

```python
country_val = final_df.Country.value_counts().values
```

```python
country_name = final_df.Country.value_counts().index
```

```python
## Pie Chart
explode = (0, 0.1, 0.3 , 0.3)
fig = plt.gcf()
fig.set_size_inches(8,8)
colors = ['#C0C0C0','#65b3ff','#99ff99','#ffcc99']
plt.pie(country_val[:4], labels = country_name[:4],explode = explode ,shadow = True,colors = colors,autopct="'2%1.2f%%'",pctdistance = 0.85)
plt.axis('equal')
centre_circle = plt.Circle((0,0),0.70,fc='gray')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```

Observation = Zomato maximum records or transaction are from India
After that USA and then UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color' , 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

## Obersvation 

1. When rating is  between 4.5 - 4.9 -> Excellent
2. When rating is between 4.0 - 3.4 -> It's very good/not bad
3. When rating is between 3.5 to 3.9 -> Good
4. When rating is between 3.0 - 3.4 -> Average
5. When rating is between 2.5 to 3.9 -> Average
6. When rating is between 2.0 to 2.4 -> Poor

Note: When rating is 0.0, it means any person has not given rating

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12 , 6)

sns.barplot(x="Aggregate rating",y="Rating Count",data = ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue = 'Rating color',data = ratings, palette = ['white','red','orange','yellow','green','green'])
```

    Obersvation:
        
    1. Not rated count is very high
    2. Maximum number of rating are between 2.4 to 3.4


```python
## Count plot

sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```
